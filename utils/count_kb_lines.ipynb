```
import re
import ast
import math
import statistics

original_directory = 'Data/KVR/train_regex_kb_indices_edited.txt'
regex_directory = 'Data/KVR/train_regex_filtered.txt'
model_directory = 'Data/KVR/train_final.txt'

def count_kb_lines(source_file):
    with open(source_file,'r') as reader:
        source_text=reader.read()

        num_overall_kb_lines=0
        num_weather_kb_lines=0
        num_navigation_kb_lines=0
        num_schedule_kb_lines=0
        
        split_dialogue_regex= re.compile('\n\n')
        split_line_regex= re.compile('\n')

        dialogues= split_dialogue_regex.split(source_text)

        for dialogue in dialogues:
            task = None
            if dialogue:
                lines = split_line_regex.split(dialogue)
                for line in lines:
                    line=line.strip()
                    if line:
                        
                        if '#' in line:
                            task= line
                            continue
                        line_indicator, _sentence = line.split(' ',1)
                        if line_indicator == '0':
                            if task == '#navigate#':
                                num_navigation_kb_lines+=1
                                num_overall_kb_lines+=1
                            elif task == '#weather#':
                                num_weather_kb_lines+=1
                                num_overall_kb_lines+=1
                            elif task == '#schedule#':
                                num_schedule_kb_lines+=1
                                num_overall_kb_lines+=1
                            continue
                    


        return num_overall_kb_lines,num_navigation_kb_lines,num_schedule_kb_lines,num_weather_kb_lines

overall_kb_lines, navigation_kb_lines, schedule_kb_lines,weather_kb_lines=count_kb_lines(original_directory)
print('original overall: {}, navigation: {}, schedule: {}, weather: {}'.format(
    overall_kb_lines,navigation_kb_lines,schedule_kb_lines,weather_kb_lines))

overall_kb_lines, navigation_kb_lines, schedule_kb_lines,weather_kb_lines=count_kb_lines(model_directory)
print('model overall: {}, navigation: {}, schedule: {}, weather: {}'.format(
    overall_kb_lines,navigation_kb_lines,schedule_kb_lines,weather_kb_lines))

overall_kb_lines, navigation_kb_lines, schedule_kb_lines,weather_kb_lines=count_kb_lines(regex_directory)
print('regex overall: {}, navigation: {}, schedule: {}, weather: {}'.format(
    overall_kb_lines,navigation_kb_lines,schedule_kb_lines,weather_kb_lines))

```
